```python
import cairo
import numpy
import random
```

```python
# A simple function to display an image in an ipython notebook
def nbimage( data ):
    from IPython.display import display, Image
    from PIL.Image import fromarray
    from StringIO import StringIO
 
    s = StringIO()
    fromarray( data ).save( s, 'png' )
    display( Image( s.getvalue() ) )
 
WIDTH = 512
HEIGHT = 288
```

```python
# this is a numpy buffer to hold the image data
data = numpy.zeros( (HEIGHT,WIDTH,4), dtype=numpy.uint8 )
 
# this creates a cairo context based on the numpy buffer
ims = cairo.ImageSurface.create_for_data( data, cairo.FORMAT_ARGB32, WIDTH, HEIGHT )
cr = cairo.Context( ims )

# VERY USEFUL DOCUMENTATION FOR UNDERSTANDING CAIRO: 
    # http://www.cairographics.org/manual/cairo-Paths.html#cairo-line-to
    
for x in range(0, 100):
    
    randomColorA = random.random()
    randomColorB = random.random()
    randomColorC = random.random()
    randomColorD = random.random()
    
    randWidthSeed = random.random()
    randomWidth = random.random() * randWidthSeed * 100
    randomXCoord = random.random() * WIDTH
    randomYCoord = random.random() * HEIGHT
    randomX2Coord = random.random() * WIDTH
    randomY2Coord = random.random() * HEIGHT
    
    curvex1 = random.random() * WIDTH
    curvex2 = random.random() * WIDTH
    curvex3 = random.random() * WIDTH
    curvey1 = random.random() * HEIGHT
    curvey2 = random.random() * HEIGHT
    curvey3 = random.random() * HEIGHT

    # draw a blue curve
    cr.set_source_rgba( randomColorA, randomColorB, randomColorC, randomColorD ) # 1-0
    cr.set_line_width( randomWidth) 
    cr.move_to( randomXCoord, randomYCoord )
    cr.curve_to(curvex1, curvey1, curvex2, curvey2, curvex3, curvey3)
    cr.stroke()

    
# display the image
nbimage( data )
```
